```python
import pandas as pd
import datetime as dt
```

```python
path = 'qp_data_result.csv'

df = pd.read_csv(path, low_memory=False)
```

```python
# оставляем только классические игры
game_names = ['Квиз, плиз! SPB', '[новички] SPB', 'Квиз, плиз!  SPB']

df = df.rename(columns={'Название игры': 'game_name'})
df = df.query('game_name in @game_names')

df.loc[df['game_name'] == 'Квиз, плиз!  SPB', 'game_name'] = 'Квиз, плиз! SPB'

df.info()
```

```python
# удаляем столбцы, где 0 значений и ранги
(
    df.drop(columns=[
        'Unnamed: 0',
        'Ранг',
        'Факультет',
        'Город',
        'Unnamed: 1',
        '8 раунд',
        '9 раунд'
    ], inplace=True
    )
)
```

```python
error_columns = ['Unnamed: 2', 'Unnamed: 0.1', 'Команда', 'Итого']
right_columns = ['Название команды', 'Место', 'Название команды', 'итого']

for i in zip(error_columns, right_columns):
    df.loc[~df[i[0]].isnull(), i[1]] = df[i[0]]
    
df.drop(columns=error_columns, inplace=True)    
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.info()
```

```python
columns_name = {'Место': 'place',
              'Название команды': 'team_name',
              '1 раунд': 'round_1',
              '2 раунд': 'round_2',
              '3 раунд': 'round_3',
              '4 раунд': 'round_4',
              '5 раунд': 'round_5',
              '6 раунд': 'round_6',
              '7 раунд': 'round_7',
              'итого': 'total',
              'Номер игры': 'game_number',
              'Название площадки': 'cafe_name',
              'Адрес площадки': 'cafe_adress',
              'Дата проведения': 'date',
              'Ссылка на сайт': 'url'}

df.rename(columns=columns_name, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
url_list = list(df[df['round_1'] > 6]['url'].unique())
round_list = ['round_1', 'round_2', 'round_3', 'round_4', 'round_5', 'round_6', 'round_7', 'total']


for url in url_list:
    x = df[df['url'] == url]['round_1']
    j = 0    
    for i in round_list:
        df.loc[df['url'] == url, i] = df[df['url'] == url][round_list[j+1]]
        j+=1
        
        if j == 7:
            break
        
    df.loc[df['url'] == url, 'total'] = x    
```

```python
round_3_mask = df['round_3'] < 7
round_5_mask = df['round_5'] < 7
round_6_mask = df['round_6'] < 7
round_7_mask = df['round_7'] < 19

df = df.loc[round_3_mask & round_5_mask & round_6_mask & round_7_mask]

round_4_mask = df['round_4'] > -6
round_5_mask = df['round_5'] >= 0
round_6_mask = df['round_6'] >= 0

df = df.loc[round_4_mask & round_5_mask & round_6_mask]

df.reset_index(drop=True, inplace=True)
```

```python
df['game_number'].str.contains('#').sum()
```

```python
df['game_number'] = df['game_number'].str.split('#', expand=True)[1]
```

```python
df.head()
```

```python
split_date = df['date'].str.split(',', expand=True)[0].str.split(' ', expand=True)

df['day'] = split_date[0]
df['month'] = split_date[1]
```

```python
month = list(df['month'].unique())
month_number = [2, 1, 12, 11, 10, 9, 8, 7, 6, 5, 4, 3]

for i in zip(month, month_number):
    df.loc[df['month'] == i[0], 'month'] = i[1]
```

```python
def create_date(row):
    row['date'] = dt.datetime(row['year'], row['month'], row['day']).date()
    return row
```

```python
df = df.apply(create_date, axis=1)
```

```python
# неправильный порядок мест

url_rand = ['https://quizplease.ru/game-page?id=35415', 'https://quizplease.ru/game-page?id=26800']

for i in url_rand:
    
    x = df.loc[df['url'] == i]
    x = x.sort_values('total', ascending=False) 
    x['place'] = x['total'].rank(ascending=False, method='first')
    x['place'] = x['place'].astype('int16')
    df.loc[df['url'] == i] = x
```
